```python
import csv
import numpy as np
import pandas as pd
import pprint
```

```python
with open('ds_salaries.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    all_salaries = []
    all_job_titles = []

    for row in csv_reader:
        salary = row['salary_in_usd']
        salary = float(salary)

        job_title = row['job_title']
        all_salaries.append(salary)
        all_job_titles.append(job_title)

    
unique_elements = set(all_job_titles)
pprint.pprint(unique_elements)
print(all_salaries)
```

```python
salary_running_sum = {}
salary_count_per_job = {}

for salary, title in zip(all_salaries, all_job_titles):
    if title not in salary_running_sum:
        salary_running_sum[title] = salary
        salary_count_per_job[title] = 1

    else:
        salary_running_sum[title] += salary
        salary_count_per_job[title] += 1


average_salary = {}
print(salary_running_sum)
print(salary_count_per_job)

for title in salary_running_sum:
    average_salary[title] = salary_running_sum[title] / salary_count_per_job[title]

print('avg salary per job title: ')

for title, average in average_salary.items():
    print(title, ': ', average)
```

```python
df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
mean_salary = df['salary_in_usd'].mean()
print(f"mean salary is {mean_salary}")

unique_job_titles = df['job_title'].unique()

average_salary_dict = dict()

for unique_job in unique_job_titles:
    people_with_this_job = df[df['job_title'] == unique_job]
    average_salary_dict[unique_job] = people_with_this_job['salary_in_usd'].mean()

print('average salary per job: ')
pprint.pprint(average_salary_dict)
```
